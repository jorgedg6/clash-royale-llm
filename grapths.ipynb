```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. Definición de Datos ---

# Etiquetas para el eje X (LLMs) y las agrupaciones (Condiciones)
llms = ['DeepSeek-V3.1', 'Llama-3.3-70B-Instruct', 'gpt-35-turbo', 'gpt-4o', 'gpt-5-chat']
conditions = ['Full (Sin Regla)', 'Full (Con Regla)', 'Mini (Sin Regla)', 'Mini (Con Regla)', 'Semi (Con Regla)']

# Datos para el Gráfico 1: Porcentaje de Mazos Mejorados
# Cada lista corresponde a una 'condición' y sus valores para cada 'llm'
mazos_data = {
    'Full (Sin Regla)': [61.0, 50.0, 0.0, 52.6, 50.0],
    'Full (Con Regla)': [47.4, 20.0, 57.1, 47.4, 60.0],
    'Mini (Sin Regla)': [61.1, 54.5, 50.0, 52.6, 68.4],
    'Mini (Con Regla)': [60.0, 57.0, 58.8, 52.6, 60.0],
    'Semi (Con Regla)': [75.0, 25.0, 55.5, 52.6, 55.0]
}

# Datos para el Gráfico 2: Porcentaje de Validez
validez_data = {
    'Full (Sin Regla)': [90.0, 30.0, 5.0, 95.0, 100.0],
    'Full (Con Regla)': [95.0, 25.0, 70.0, 95.0, 100.0],
    'Mini (Sin Regla)': [90.0, 55.0, 40.0, 95.0, 95.0],
    'Mini (Con Regla)': [75.0, 70.0, 85.0, 95.0, 100.0],
    'Semi (Con Regla)': [100.0, 20.0, 90.0, 95.0, 100.0]
}

# --- 2. Configuración de Trazado Común ---

# Número de LLMs (grupos en el eje X)
n_llms = len(llms)
# Número de condiciones (barras por grupo)
n_conditions = len(conditions)

# Posiciones de los grupos en el eje X (p.ej., 0, 1, 2, 3, 4)
x = np.arange(n_llms)

# Ancho de cada barra individual
bar_width = 0.15

# Calcular los desplazamientos desde el centro para cada una de las 5 barras
# Esto centra las 5 barras alrededor de la marca del eje X (0, 1, 2...)
offsets = (np.arange(n_conditions) - (n_conditions - 1) / 2) * bar_width

# Colores para las 5 condiciones (puedes personalizarlos)
colors = ["#3775a1", "#ca8e5a", "#44bc44", "#d95f5f", "#aa86cb"]

# --- 3. Generar Gráfico 1: Porcentaje de Mazos Mejorados ---

fig1, ax1 = plt.subplots(figsize=(15, 8))

# Iterar sobre cada condición y trazar sus barras
for i, condition in enumerate(conditions):
    ax1.bar(
        x + offsets[i],         # Posición X (centro + desplazamiento)
        mazos_data[condition],  # Datos (altura de la barra)
        bar_width,              # Ancho de la barra
        label=condition,        # Etiqueta para la leyenda
        color=colors[i]         # Color
    )

# Configuración de etiquetas y título
ax1.set_ylabel('Porcentaje Mejorados (%)', fontsize=12)
ax1.set_title('Comparación de Porcentaje de Mazos Mejorados', fontsize=16)
ax1.set_xticks(x)
ax1.set_xticklabels(llms, rotation=45, ha='right')
ax1.set_ylim(0, 105) # Dejar espacio para la leyenda

# Añadir leyenda
ax1.legend(title='Condición del Prompt', bbox_to_anchor=(1.02, 1), loc='upper left')

# Añadir rejilla para facilitar la lectura
ax1.yaxis.grid(True, linestyle='--', alpha=0.7)
ax1.set_axisbelow(True) # Poner la rejilla detrás de las barras

# Ajustar layout para que quepa todo
plt.tight_layout()

# Guardar la imagen
plt.savefig('grafico_mazos_mejorados.png')
print("Gráfico 'grafico_mazos_mejorados.png' guardado.")

# --- 4. Generar Gráfico 2: Porcentaje de Validez ---

fig2, ax2 = plt.subplots(figsize=(15, 8))

# Iterar sobre cada condición y trazar sus barras
for i, condition in enumerate(conditions):
    ax2.bar(
        x + offsets[i],             # Posición X
        validez_data[condition],    # Datos
        bar_width,                  # Ancho
        label=condition,            # Etiqueta
        color=colors[i]             # Color
    )

# Configuración de etiquetas y título
ax2.set_ylabel('Porcentaje de Validez (%)', fontsize=12)
ax2.set_title('Comparación de Porcentaje de Validez', fontsize=16)
ax2.set_xticks(x)
ax2.set_xticklabels(llms, rotation=45, ha='right')
ax2.set_ylim(0, 105) # Límite Y hasta 105%

# Añadir leyenda
ax2.legend(title='Condición del Prompt', bbox_to_anchor=(1.02, 1), loc='upper left')

# Añadir rejilla
ax2.yaxis.grid(True, linestyle='--', alpha=0.7)
ax2.set_axisbelow(True)

# Ajustar layout
plt.tight_layout()

# Guardar la imagen
plt.savefig('grafico_validez.png')
print("Gráfico 'grafico_validez.png' guardado.")
```
